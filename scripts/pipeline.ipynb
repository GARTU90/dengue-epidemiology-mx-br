```python
import pandas as pd
import numpy as np
```

```python
M =  pd.read_csv("data/dengue_mexico.csv")
```

```python
B = pd.read_csv("data/dengue_brasil.csv",low_memory=False)
B.info()

```

```python
C2 = """SEXO	CS_SEXO
EDAD_ANOS	EDAD
ENTIDAD_RES	SG_UF
MUNICIPIO_RES	ID_MN_RESI
ENTIDAD_UM_NOTIF	SG_UF_NOT
MUNICIPIO_UM_NOTIF	ID_MUNICIP
INSTITUCION_UM_NOTIF	ID_UNIDADE
FECHA_SIGN_SINTOMAS	DT_SIN_PRI
DIABETES	DIABETES
HIPERTENSION	HIPERTENSA
ENFERMEDAD_ULC_PEPTICA	ACIDO_PEPT
ENFERMEDAD_RENAL	RENAL
INMUNOSUPR	AUTO_IMUNE
CIRROSIS_HEPATICA	HEPATOPAT
EMBARAZO	CS_GESTANT
DEFUNCION	DEFUNCION
RESULTADO_PCR	SOROTIPO
ESTATUS_CASO	CLASSI_FIN
TOMA_MUESTRA	TOMA_MUESTRA
HEMORRAGICOS	ALRM_SANG"""

C = [r.split("\t") for r in C2.strip().split("\n")]
C1 = {key:value for key,value in C}
C2 = {key:value for value,key in C}
C1.values()
```

```python
T = M[C1.keys()].copy()
T.info()

#podemos observar que no hay nulos
```

```python
S = []
I = []
F = []

for c in B.columns:
  t = B[c].dtype
  if t == 'float64':F.append(c)
  elif t == 'object':S.append(c)
  elif t == 'int64':I.append(c)

#print(f"S: {S}\nI: {I}\nF: {F}")

MS = [C2[s] for s in S]
MI = [C2[s] for s in I]
MF = [C2[s] for s in F]
#print(f"MS: {MS}")
#print(f"MI: {MI}")
#print(f"MF: {MF}")
```

```python
T[MF] = T[MF].astype("float64")
T[MI] = T[MI].astype("int64")
T[MS] = T[MS].astype("object")
```

```python
T["SEXO"]=T["SEXO"].replace({1:"F",2:"M"})
T["DEFUNCION"]=T["DEFUNCION"].replace({2:0})
T["TOMA_MUESTRA"]=T["TOMA_MUESTRA"].replace({4:1.0})
```

```python
B["CS_GESTANT"] = B["CS_GESTANT"].replace({1:"1.0",2:"1.0",3:"1.0",4:"2.0",5:"2.0",6:"2.0",9:"2.0",np.nan:"2.0"})
```

```python
for c in B.columns:
  #print(c)
  print(B[c].value_counts(dropna=False))
```

```python
#correcion de nans y valores que no deberian estar como los 9s


B["TOMA_MUESTRA"] = B["TOMA_MUESTRA"].fillna(0)
B["CS_SEXO"] = B["CS_SEXO"].fillna("I")
B["DIABETES"] = B["DIABETES"].replace({np.nan:0,9.0:0})
B["HEPATOPAT"] = B["HEPATOPAT"].replace({np.nan:0,9.0:0})
B["RENAL"] = B["RENAL"].replace({np.nan:0,9.0:0})
B["HIPERTENSA"] = B["HIPERTENSA"].replace({np.nan:0,9.0:0})
B["ACIDO_PEPT"] = B["ACIDO_PEPT"].replace({np.nan:0,9.0:0})
B["AUTO_IMUNE"] = B["AUTO_IMUNE"].replace({np.nan:0,9.0:0})
B["ALRM_SANG"] = B["ALRM_SANG"].replace({np.nan:0,9.0:0})
B["CLASSI_FIN"] = B["CLASSI_FIN"].replace({np.nan:1})
```

```python
B.info()
```

```python
T.columns = C1.values()
```

```python
T.insert(0,"PAIS","Mexico")
B.insert(0,"PAIS","Brasil")
```

```python
N = pd.concat([T,B],ignore_index=True)
N
```

```python
N.info()
```

```python
for c in N.columns:
  #print(c)
  print(N[c].value_counts(dropna=False))
```

```python
N.to_csv("data/Dengue_BiNacional.csv",index=False)
```
